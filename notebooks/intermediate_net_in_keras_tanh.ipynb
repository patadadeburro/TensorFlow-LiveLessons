# Intermediate Neural Network in Keras

In this notebook, we improve our [introductory shallow net](https://github.com/the-deep-learners/TensorFlow-LiveLessons/blob/master/notebooks/shallow_net_in_keras.ipynb) from Lesson 1 by applying the theory we have covered since. 

#### Set seed for reproducibility

```python
import numpy as np
np.random.seed(42)
```

#### Load dependencies

```python
import keras
from   keras.datasets   import mnist
from   keras.models     import Sequential
from   keras.layers     import Dense
from   keras.optimizers import SGD
```

#### Load data

```python
#(X_train, y_train), (X_test, y_test) = mnist.load_data()
(X_train, y_train), (X_test, y_test) = mnist.load_data( 'C:/aat/data/mnist_handwritten_digits' )

# (X_train, y_train), (X_test, y_test) = mnist.load_data( 'C:/aat/data/mnist_handwritten_digits' )
```

#### Preprocess data

```python
X_train = X_train.reshape(60000, 784).astype('float32')
X_test = X_test.reshape(10000, 784).astype('float32')
```

```python
X_train /= 255
X_test /= 255
```

```python
n_classes = 10
y_train = keras.utils.to_categorical(y_train, n_classes)
y_test = keras.utils.to_categorical(y_test, n_classes)
```

#### Design neural network architecture

'''tanh was not very good'''
model = Sequential()
model.add(Dense(64, activation='tanh', input_shape=(784,)))
model.add(Dense(10, activation='softmax'))

```python
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(784,)))
model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

```python
model.summary()
```

#### Configure model

```python
#model.compile(loss='categorical_crossentropy', optimizer=SGD(lr=0.1), metrics=['accuracy'])
model.compile(loss='categorical_crossentropy', optimizer=SGD(lr=0.1), metrics=['accuracy'])
```

#### Train!

```python
model.fit(X_train, y_train, batch_size=128, epochs=200, verbose=1, validation_data=(X_test, y_test))
```
